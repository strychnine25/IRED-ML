# Data analysis
Performing similar data exploration as was done for the IRED dataset earlier in the project

```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
sns.set(rc={'figure.figsize': (5, 5)})
sns.set(font_scale=1.5)
sns.set_style('white')
sns.set_context("poster")
sns.set_theme()
%config InlineBackend.figure_format = 'retina'
```

```python
df = pd.read_excel('Prelim and active site data.xlsx', engine='openpyxl')
df.dropna(subset=['Enantiomer'], inplace=True)
df.head(2)
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(data=df, x='ActiveSiteCharge', hue='Enantiomer')
plt.xticks(rotation='horizontal');
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(data=df, x='NumOfAcidicRes', hue='Enantiomer')
plt.xticks(rotation='horizontal');
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(data=df, x='NumOfBasicRes', hue='Enantiomer')
plt.xticks(rotation='horizontal');
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(data=df, x='NumOfHisRes', hue='Enantiomer')
plt.xticks(rotation='horizontal');
```

```python
# Finding longest and shortest sequence lengths of active site sequences
as_seq_length = []
for i in df['ActiveSiteSequence']:
    a = len(i)
    as_seq_length.append(a)
    as_seq_length.sort()
#as_seq_length
```

```python
# Finding longest and shortest sequence lengths of cofactor site sequences
cf_seq_length = []
for i in df['CofactorSiteSequence']:
    a = len(i)
    cf_seq_length.append(a)
    cf_seq_length.sort()
#cf_seq_length
```

```python
# Finding longest and shortest sequence lengths of full sequences
seq_length = []
for i in df['Sequence']:
    a = len(i)
    seq_length.append(a)
    seq_length.sort()
seq_length
```
